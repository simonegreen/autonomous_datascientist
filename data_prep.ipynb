```python
# Import data csv file
import pandas as pd
import numpy as np
df3 = pd.read_csv(".csv")

print("Starting Columns:", str(df3.shape[1]))
df3.dropna(axis=1, inplace=True) # remove columns with NaN values
#print(df.head(10))
print("Columns with Values:", str(df3.shape[1]))

df3.head(10)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

'''Converts the qualitative column col in df to quantitative values'''
def qual_to_quant(df, col):
  vals = df[col].unique()
  num_vals = range(df[col].nunique())
  mapping = {k:v for k,v in zip(vals, num_vals)}
  df[col] = df[col].apply(lambda x: mapping[x])

"Returns an array with the indexes of the top n values in arr"
def get_top_n_idx(n, arr):
  arr = np.abs(arr)
  top = np.argpartition(arr, -n)[-n:]
  return top


'''Prints the selected features from data using PCA.
Optionally, can return the final selected features as a np array.

PCA: using 5 components
Feature Selection: getting the top two features contributing to each of the 5 components'''
def get_features(data):
  copy = data.copy(deep = True)
  feat_options = copy.iloc[:, 2:].copy(deep = True) # don't scale ip/device name
  scaler = StandardScaler()
  scaler.fit(feat_options)
  scaled_data = scaler.transform(feat_options)
  pca = PCA(n_components=5)
  pca.fit(scaled_data)
  #print(pca.components_[0])
  n_pcs= pca.components_.shape[0]

  # get the index of the most important feature on EACH component
  #most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
  #most_important = [sorted(np.abs(pca.components_[i]))[:2] ]
  most_important = []
  for i in range(n_pcs):
    top = get_top_n_idx(10//n_pcs, pca.components_[i]) # for 5 components, gets 10//5 = top 2
    # print(top)
    # print(most_important)
    most_important = most_important + list(top)

  most_important_names = [feat_options.columns[most_important[i]] for i in range(10)]

  print("Most Important Idx:", most_important)
  print("Most Important Features:", most_important_names)
  # arr = np.array(pca.components_)
  # abs_arr = np.abs(arr)
  # print(abs_arr[0])
  # top = np.argmax(abs_arr[0])
  # print(top)
  unique_feats = np.unique(most_important_names)
  print("Unique Features:", unique_feats)
  # return unique_feats
  return
```
