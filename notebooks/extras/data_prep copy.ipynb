```python
import pandas as pd
import numpy as np

df = pd.read_csv("combined_no_aggregation.csv")
print("Starting Columns:", str(df.shape[1]))
df.dropna(axis=1, inplace=True) # remove columns with NaN values
#print(df.head(10))
print("Columns with Values:", str(df.shape[1]))
```

```python
def qual_to_quant(df, col):
  vals = df[col].unique()
  num_vals = range(df[col].nunique())
  mapping = {k:v for k,v in zip(vals, num_vals)}
  df[col] = df[col].apply(lambda x: mapping[x])
```

```python
MAIN = ['device_name_main', 'source_ip_main']
num_entries = df.shape[0]
drop = []
for i in df:
  if df[i].dtype == 'O' and i not in MAIN: # qualitative
      qual_to_quant(df, i)
      # TO-DO: we don't have to do this, but might add efficiency
      if df[i].nunique() > (num_entries / 2) or df[i].nunique() == 1: # if more than 1/2 of data points have a unique label, drop the column. can change this!
         drop.append(i)
print(drop)
cleaned_df = df.drop(columns = drop)
print("Final Columns:", str(cleaned_df.shape[1]))
df.to_csv('cleaned_no_aggregation.csv', index=False)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def get_features(data):
  copy = data.copy(deep = True)
  feat_options = copy.iloc[:, 2:].copy(deep = True) # don't scale ip/device name
  scaler = StandardScaler()
  scaler.fit(feat_options)
  scaled_data = scaler.transform(feat_options)
  pca = PCA(n_components=5)
  pca.fit(scaled_data)
  #print(pca.components_[0])
  n_pcs= pca.components_.shape[0]

  # get the index of the most important feature on EACH component
  #most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
  most_important = [sorted(np.abs(pca.components_[i]))[:2] ]

  most_important_names = [feat_options.columns[most_important[i]] for i in range(n_pcs)]

  #print(most_important)
  #print(most_important_names)
  arr = np.array(pca.components_)
  abs_arr = np.abs(arr)
  # print(abs_arr[0])
  # top = np.argmax(abs_arr[0])
  # print(top)
  unique_feats = np.unique(most_important_names)
  return(unique_feats)
```
